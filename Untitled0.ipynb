```python
import random

numero_persona = int(input("Dame un número (1, 2 o 3): "))
numero_maquina = random.randint(1, 3)

print("El número de la persona es:", numero_persona)
print("El número de la máquina es:", numero_maquina)

# Verificación de par o impar para el número de la persona
if numero_persona % 2 == 0:
    print(f"El número de la persona ({numero_persona}) es par.")
else:
    print(f"El número de la persona ({numero_persona}) es impar.")

# Verificación de par o impar para el número de la máquina
if numero_maquina % 2 == 0:
    print(f"El número de la máquina ({numero_maquina}) es par.")
else:
    print(f"El número de la máquina ({numero_maquina}) es impar.")

# Lógica del juego
if numero_persona == numero_maquina:
    print("Ha ganado la persona!")
else:
    print("Ha ganado la máquina.")
```

```python
import codecs

def cifrar_rot13(mensaje):
    return codecs.encode(mensaje, 'rot_13')

def descifrar_rot13(mensaje_cifrado):
    return codecs.decode(mensaje_cifrado, 'rot_13')

# Ejemplo de uso
mensaje = input("Introduce el mensaje a cifrar: ")

# Cifrado
mensaje_cifrado = cifrar_rot13(mensaje)
print("Mensaje cifrado con ROT13:", mensaje_cifrado)

# Descifrado
mensaje_descifrado = descifrar_rot13(mensaje_cifrado)
print("Mensaje descifrado con ROT13:", mensaje_descifrado)
```
